```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
Y = []

for line in open('data_poly.csv'):
    x, y = line.split(',')
    x = float(x)
    X.append([1, x, x*x])
    Y.append(float(y))

X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X[:, 1], Y)
plt.show()
```

```python
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
y_hat = np.dot(X, w)
```

```python
plt.scatter(X[:, 1], Y)
plt.plot(sorted(X[:, 1]), sorted(y_hat), color="red")
plt.show()
```

```python
d1 = Y - y_hat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print(r2)
```
