```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the data:
X = []
Y = []

file = open("data_1d.csv")

for line in file:
    x, y = line.split(",")
    X.append(float(x))
    Y.append(float(y))

# Convert into numpy's array:
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
# Calculate a and b:
denominator = X.dot(X) - X.mean() * X.sum()
a = (X.dot(Y) - Y.mean() * X.sum()) / denominator
b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y)) / denominator
```

```python
# predicted Y:
Y_hat = a * X + b
```

```python
# Plot everything:
plt.scatter(X, Y)
plt.plot(X, Y_hat, color="red")
plt.show()
```

```python
# Calculate the r-squared:
ss_res = (Y - Y_hat).dot(Y - Y_hat)
ss_tot = (Y - Y.mean()).dot(Y - Y.mean())
r_squared = 1 - (ss_res / ss_tot)
print(f"the r-squared is: {r_squared}")
```
