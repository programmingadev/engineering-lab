```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
Y = []

file = open('data_2d.csv', 'r')

for lines in file:
    x1, x2, y = lines.split(',')
    X.append([float(x1), float(x2), 1])
    Y.append(float(y))

file.close()

X = np.array(X)
Y = np.array(Y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:, 0], X[:, 1], Y)
plt.show()
```

```python
#  Calculate weights:
w = np.linalg.solve(np.dot(X.T, X), np.dot(X.T, Y))
y_hat = np.dot(X,w)

#  Compute the r-squared:
d1 = Y - y_hat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print(f"R-squared: {r2}")
```
